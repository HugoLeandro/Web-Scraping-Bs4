```python
from bs4 import BeautifulSoup
import requests
import re


link = "https://pypi.org/project/beautifulsoup4/"
requisicao = requests.get(link)
            
```

```python
site = BeautifulSoup(requisicao.text, "html.parser")
print(site.prettify())
```

```python
tbody = site.find("tbody")
print(tbody.prettify())
```

```python
linhas = tbody.find_all("tr")
for linha in linhas:
    try:
        nome = linha.find(class_="ePTNty").text
        codigo = linha.find(class_="coin-item-symbol").text
        valores = linha.find_all(string=re.compile("\$"))
        preco = valores[0]
        percentuais = linha.find_all(string=re.compile("%"))
        vari_1h = ""
        vari_24H = ""
        vari_7d = ""
        marketcap = valores[2]
        volume = valores[3]
        dic = {"nome": nome, "codigo": codigo, "valores": valores, "preco": preco, "vari_1h": vari_1h, "vari24h": vari_24H, "vari_7d": vari_7d, }

    except AttributeError:
        break
```
